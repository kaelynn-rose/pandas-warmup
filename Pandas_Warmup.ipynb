One way to approach this problem is to first create a custom function that returns a boolean depending on whether a given string is in the title, pass that function to the apply method (which applies the function to each row in the job title column), and then sum up the number of True values.
    
```python    
def string_job_title(title,string):
    if string.lower() in title.lower():
        return True
    else:
        return False

sum(salary['JobTitle'].apply(lambda x: string_job_title(x,'General')))
``` 
